```python
import librosa #python package for audio and music analysis
import torch #ml library
import IPython.display as display  #interactive command line for terminal
from transformers import Wav2Vec2ForCTC, Wav2Vec2Tokenizer  #[how find tune in speech recogniser, tokenizer]
import numpy as np #for arrays
```

```python
tokenizer= Wav2Vec2Tokenizer.from_pretrained("facebook/wav2vec2-base-960h") #pretrined model is import
model= Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-base-960h")
```

```python
audio, sampling_rate=librosa.load("v.mp3",sr=16000)
```

```python
audio,sampling_rate
```

```python
display.Audio("v.mp3",autoplay=True) #plays audio
```

```python
input_values= tokenizer(audio,return_tensors='pt').input_values
input_values
```

```python
logits = model(input_values).logits #non normalized predictions
logits
```

```python
predicted_ids=torch.argmax(logits,dim=-1)
```

```python
transcriptions=tokenizer.decode(predicted_ids[0])
```
